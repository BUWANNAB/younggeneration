```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# 设置中文字体
plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]
plt.rcParams["axes.unicode_minus"] = False  # 解决负号显示问题

# 生成示例数据
np.random.seed(42)
n_points = 50

# 方案1：房价与面积关系
house_data = pd.DataFrame({
    '面积(平方米)': np.random.normal(100, 30, n_points),
    '价格(万元)': np.random.normal(200, 80, n_points) + np.random.normal(1, 0.5, n_points) * 1.5,
    '区域': np.random.choice(['市中心', '郊区', '近郊区'], n_points)
})

# 方案2：考试成绩分析
exam_data = pd.DataFrame({
    '数学成绩': np.random.normal(70, 15, n_points),
    '语文成绩': np.random.normal(75, 10, n_points) + np.random.normal(0, 8, n_points),
    '性别': np.random.choice(['男', '女'], n_points)
})

# 方案3：股票收益与风险
stock_data = pd.DataFrame({
    '风险(标准差)': np.random.uniform(10, 30, n_points),
    '收益(年化%)': np.random.uniform(5, 15, n_points) - np.random.uniform(0, 5, n_points) * 0.2,
    '行业': np.random.choice(['科技', '金融', '消费', '医药'], n_points)
})

# 绘制方案1：房价与面积关系
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    house_data['面积(平方米)'], 
    house_data['价格(万元)'], 
    c=house_data['区域'].map({'市中心': 'red', '近郊区': 'blue', '郊区': 'green'}),
    alpha=0.7, 
    s=50,
    edgecolors='w', 
    linewidth=1
)

# 添加标题和标签
plt.title('房屋面积与价格关系散点图')
plt.xlabel('面积(平方米)')
plt.ylabel('价格(万元)')

# 添加图例
legend = plt.legend(
    handles=scatter.legend_elements()[0], 
    labels=['市中心', '近郊区', '郊区'],
    title='区域'
)
plt.gca().add_artist(legend)

# 添加网格线
plt.grid(True, linestyle='--', alpha=0.7)

# 方案2：考试成绩分析
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='数学成绩', 
    y='语文成绩', 
    hue='性别',
    palette={'男': 'blue', '女': 'pink'},
    s=80,
    alpha=0.8,
    data=exam_data
)

plt.title('学生数学与语文成绩散点图')
plt.xlabel('数学成绩')
plt.ylabel('语文成绩')
plt.grid(True, linestyle='--', alpha=0.7)

# 添加回归线
sns.regplot(
    x='数学成绩', 
    y='语文成绩', 
    data=exam_data, 
    scatter=False, 
    color='black', 
    line_kws={'linestyle': '--', 'alpha': 0.5}
)

# 方案3：股票收益与风险
plt.figure(figsize=(12, 7))
sns.scatterplot(
    x='风险(标准差)', 
    y='收益(年化%)', 
    hue='行业',
    palette='viridis',
    s=100,
    alpha=0.8,
    data=stock_data
)

plt.title('股票风险与收益关系散点图')
plt.xlabel('风险(标准差)')
plt.ylabel('收益(年化%)')
plt.grid(True, linestyle='--', alpha=0.7)

# 添加参考线
plt.axhline(y=stock_data['收益(年化%)'].mean(), color='r', linestyle='--', alpha=0.5, label='平均收益')
plt.axvline(x=stock_data['风险(标准差)'].mean(), color='g', linestyle='--', alpha=0.5, label='平均风险')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd
import matplotlib.colors as mcolors

# 设置中文字体
plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]
plt.rcParams["axes.unicode_minus"] = False  # 解决负号显示问题

# 方案1：城市事件影响力涟漪图
cities = {
    '北京': (116.4074, 39.9042),
    '上海': (121.4737, 31.2304),
    '广州': (113.2644, 23.1291),
    '深圳': (114.0579, 22.5431),
    '成都': (104.0658, 30.5728),
    '武汉': (114.3055, 30.5928),
    '西安': (108.9480, 34.2631)
}

# 模拟事件数据（事件发生时间、中心城市、强度）
events = [
    {'time': 0, 'city': '北京', 'intensity': 8},
    {'time': 10, 'city': '上海', 'intensity': 6},
    {'time': 20, 'city': '广州', 'intensity': 7},
    {'time': 30, 'city': '成都', 'intensity': 5},
    {'time': 40, 'city': '武汉', 'intensity': 9}
]

# 方案2：社交网络热点传播涟漪图
# 生成社交网络节点数据
np.random.seed(42)
n_nodes = 30
nodes = pd.DataFrame({
    'x': np.random.rand(n_nodes),
    'y': np.random.rand(n_nodes),
    'influence': np.random.uniform(1, 5, n_nodes),  # 节点影响力
    'category': np.random.choice(['A', 'B', 'C', 'D'], n_nodes)
})

# 模拟热点事件
hotspots = [
    {'time': 0, 'node_id': 5, 'intensity': 10},
    {'time': 15, 'node_id': 12, 'intensity': 8},
    {'time': 30, 'node_id': 22, 'intensity': 7}
]

# 方案3：金融市场波动涟漪图
# 生成金融市场数据
stocks = {
    '股票A': (0.2, 0.3),
    '股票B': (0.4, 0.1),
    '股票C': (0.6, 0.2),
    '股票D': (0.8, 0.4),
    '股票E': (0.3, 0.6),
    '股票F': (0.5, 0.8),
    '股票G': (0.7, 0.5),
    '股票H': (0.1, 0.7)
}

# 模拟市场事件
market_events = [
    {'time': 0, 'stock': '股票A', 'intensity': 9, 'color': 'red'},
    {'time': 12, 'stock': '股票E', 'intensity': 7, 'color': 'blue'},
    {'time': 25, 'stock': '股票G', 'intensity': 8, 'color': 'green'},
    {'time': 38, 'stock': '股票C', 'intensity': 6, 'color': 'purple'}
]

# 方案1：城市事件影响力涟漪图实现
fig1, ax1 = plt.subplots(figsize=(10, 8))

# 设置地图边界
ax1.set_xlim(100, 130)
ax1.set_ylim(20, 45)

# 绘制城市点
city_points = []
for city, (lon, lat) in cities.items():
    point, = ax1.plot(lon, lat, 'o', markersize=8, color='black', label=city)
    city_points.append(point)
    ax1.annotate(city, (lon, lat), xytext=(5, 5), textcoords='offset points')

# 设置标题和标签
ax1.set_title('城市事件影响力传播涟漪图')
ax1.set_xlabel('经度')
ax1.set_ylabel('纬度')

# 网格线
ax1.grid(True, linestyle='--', alpha=0.5)

# 创建涟漪对象列表
ripples = []

def init1():
    return []

def update1(frame):
    # 清除旧的涟漪
    for ripple in ripples[:]:
        if ripple[0].get_alpha() <= 0:
            for artist in ripple:
                artist.remove()
            ripples.remove(ripple)
    
    # 检查是否有新事件
    for event in events:
        if frame >= event['time'] and frame < event['time'] + 40:
            elapsed = frame - event['time']
            if elapsed % 5 == 0:  # 每5帧创建一个新的涟漪
                lon, lat = cities[event['city']]
                radius = elapsed / 4
                alpha = 0.8 - (elapsed / 40) * 0.8
                
                # 创建涟漪圆圈
                circle1 = plt.Circle((lon, lat), radius, fill=False, color='red', alpha=alpha, linewidth=2)
                circle2 = plt.Circle((lon, lat), radius*0.7, fill=False, color='red', alpha=alpha*0.7, linewidth=1.5)
                circle3 = plt.Circle((lon, lat), radius*0.4, fill=False, color='red', alpha=alpha*0.4, linewidth=1)
                
                ax1.add_artist(circle1)
                ax1.add_artist(circle2)
                ax1.add_artist(circle3)
                
                ripples.append([circle1, circle2, circle3])
    
    return [artist for ripple in ripples for artist in ripple]

# 创建动画
ani1 = FuncAnimation(fig1, update1, frames=np.arange(0, 60), init_func=init1, blit=True, interval=100)

# 方案2：社交网络热点传播涟漪图实现
fig2, ax2 = plt.subplots(figsize=(10, 8))

# 绘制节点
colors = {'A': 'red', 'B': 'blue', 'C': 'green', 'D': 'purple'}
for category, color in colors.items():
    cat_nodes = nodes[nodes['category'] == category]
    ax2.scatter(cat_nodes['x'], cat_nodes['y'], c=color, s=cat_nodes['influence']*15, 
               alpha=0.7, label=category, edgecolors='w', linewidth=1)

# 设置标题和标签
ax2.set_title('社交网络热点传播涟漪图')
ax2.set_xlabel('X坐标')
ax2.set_ylabel('Y坐标')
ax2.legend(title='节点类别')

# 隐藏坐标轴刻度
ax2.set_xticks([])
ax2.set_yticks([])

# 网格线
ax2.grid(True, linestyle='--', alpha=0.3)

# 创建涟漪对象列表
social_ripples = []

def init2():
    return []

def update2(frame):
    # 清除旧的涟漪
    for ripple in social_ripples[:]:
        if ripple[0].get_alpha() <= 0:
            for artist in ripple:
                artist.remove()
            social_ripples.remove(ripple)
    
    # 检查是否有新热点
    for hotspot in hotspots:
        if frame >= hotspot['time'] and frame < hotspot['time'] + 30:
            elapsed = frame - hotspot['time']
            if elapsed % 4 == 0:  # 每4帧创建一个新的涟漪
                node = nodes.iloc[hotspot['node_id']]
                x, y = node['x'], node['y']
                intensity = hotspot['intensity']
                radius = elapsed / 5
                alpha = 0.9 - (elapsed / 30) * 0.9
                
                # 创建涟漪圆圈
                circle1 = plt.Circle((x, y), radius, fill=False, color='orange', alpha=alpha, linewidth=2)
                circle2 = plt.Circle((x, y), radius*0.7, fill=False, color='orange', alpha=alpha*0.7, linewidth=1.5)
                
                ax2.add_artist(circle1)
                ax2.add_artist(circle2)
                
                social_ripples.append([circle1, circle2])
    
    return [artist for ripple in social_ripples for artist in ripple]

# 创建动画
ani2 = FuncAnimation(fig2, update2, frames=np.arange(0, 50), init_func=init2, blit=True, interval=150)

# 方案3：金融市场波动涟漪图实现
fig3, ax3 = plt.subplots(figsize=(10, 8))

# 绘制股票点
stock_points = []
for stock, (x, y) in stocks.items():
    point, = ax3.plot(x, y, 'o', markersize=8, color='black', label=stock)
    stock_points.append(point)
    ax3.annotate(stock, (x, y), xytext=(5, 5), textcoords='offset points')

# 设置标题和标签
ax3.set_title('金融市场波动涟漪图')
ax3.set_xlabel('风险指标')
ax3.set_ylabel('收益指标')

# 网格线
ax3.grid(True, linestyle='--', alpha=0.5)

# 设置坐标轴范围
ax3.set_xlim(0, 1)
ax3.set_ylim(0, 1)

# 创建涟漪对象列表
market_ripples = []

def init3():
    return []

def update3(frame):
    # 清除旧的涟漪
    for ripple in market_ripples[:]:
        if ripple[0].get_alpha() <= 0:
            for artist in ripple:
                artist.remove()
            market_ripples.remove(ripple)
    
    # 检查是否有新市场事件
    for event in market_events:
        if frame >= event['time'] and frame < event['time'] + 35:
            elapsed = frame - event['time']
            if elapsed % 5 == 0:  # 每5帧创建一个新的涟漪
                x, y = stocks[event['stock']]
                intensity = event['intensity']
                radius = (elapsed / 35) * 0.3
                alpha = 0.8 - (elapsed / 35) * 0.8
                
                # 创建涟漪圆圈
                circle1 = plt.Circle((x, y), radius, fill=False, color=event['color'], alpha=alpha, linewidth=2)
                circle2 = plt.Circle((x, y), radius*0.6, fill=False, color=event['color'], alpha=alpha*0.7, linewidth=1.5)
                
                ax3.add_artist(circle1)
                ax3.add_artist(circle2)
                
                market_ripples.append([circle1, circle2])
    
    return [artist for ripple in market_ripples for artist in ripple]

# 创建动画
ani3 = FuncAnimation(fig3, update3, frames=np.arange(0, 50), init_func=init3, blit=True, interval=120)

plt.tight_layout()
plt.show()
```
