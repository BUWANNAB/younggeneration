```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# 
plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]
plt.rcParams["axes.unicode_minus"] = False  # 

# 
np.random.seed(42)
n_points = 50

# 1
house_data = pd.DataFrame({
    '()': np.random.normal(100, 30, n_points),
    '()': np.random.normal(200, 80, n_points) + np.random.normal(1, 0.5, n_points) * 1.5,
    '': np.random.choice(['', '', ''], n_points)
})

# 2
exam_data = pd.DataFrame({
    '': np.random.normal(70, 15, n_points),
    '': np.random.normal(75, 10, n_points) + np.random.normal(0, 8, n_points),
    '': np.random.choice(['', ''], n_points)
})

# 3
stock_data = pd.DataFrame({
    '()': np.random.uniform(10, 30, n_points),
    '(%)': np.random.uniform(5, 15, n_points) - np.random.uniform(0, 5, n_points) * 0.2,
    '': np.random.choice(['', '', '', ''], n_points)
})

# 1
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    house_data['()'], 
    house_data['()'], 
    c=house_data[''].map({'': 'red', '': 'blue', '': 'green'}),
    alpha=0.7, 
    s=50,
    edgecolors='w', 
    linewidth=1
)

# 
plt.title('')
plt.xlabel('()')
plt.ylabel('()')

# 
legend = plt.legend(
    handles=scatter.legend_elements()[0], 
    labels=['', '', ''],
    title=''
)
plt.gca().add_artist(legend)

# 
plt.grid(True, linestyle='--', alpha=0.7)

# 2
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='', 
    y='', 
    hue='',
    palette={'': 'blue', '': 'pink'},
    s=80,
    alpha=0.8,
    data=exam_data
)

plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.grid(True, linestyle='--', alpha=0.7)

# 
sns.regplot(
    x='', 
    y='', 
    data=exam_data, 
    scatter=False, 
    color='black', 
    line_kws={'linestyle': '--', 'alpha': 0.5}
)

# 3
plt.figure(figsize=(12, 7))
sns.scatterplot(
    x='()', 
    y='(%)', 
    hue='',
    palette='viridis',
    s=100,
    alpha=0.8,
    data=stock_data
)

plt.title('')
plt.xlabel('()')
plt.ylabel('(%)')
plt.grid(True, linestyle='--', alpha=0.7)

# 
plt.axhline(y=stock_data['(%)'].mean(), color='r', linestyle='--', alpha=0.5, label='')
plt.axvline(x=stock_data['()'].mean(), color='g', linestyle='--', alpha=0.5, label='')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# 
plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]
plt.rcParams["axes.unicode_minus"] = False  # 

# 1
cities = pd.DataFrame({
    '': ['', '', '', '', '', '', '', '', '', ''],
    'GDP()': [36000, 38000, 25000, 28000, 16000, 17000, 15000, 14000, 23000, 10000],
    '()': [2154, 2487, 1530, 1303, 980, 1633, 1121, 850, 3124, 1020],
    '(/)': [6.5, 6.8, 3.5, 6.2, 3.8, 2.0, 2.2, 3.2, 1.5, 1.8],
    '': ['', '', '', '', '', '', '', '', '', '']
})

# 2
cars = pd.DataFrame({
    '': ['Model 3', '', '', 'L', '', '', 'A4L', 'C', '3', ''],
    '(L/100km)': [16, 6.0, 6.3, 7.5, 5.5, 5.8, 6.5, 7.0, 6.8, 5.2],
    '(kW)': [202, 152, 162, 137, 130, 146, 185, 190, 190, 163],
    '()': [25, 18, 18, 22, 12, 19, 36, 38, 37, 23],
    '': ['', '', '', 'SUV', '', '', '', '', '', '']
})

# 3
education = pd.DataFrame({
    '': ['', '', '', '', '', '', '', '', '', ''],
    '(%)': [51.6, 88.2, 61.8, 56.5, 66.9, 50.5, 93.5, 66.0, 70.2, 72.7],
    'GDP(%)': [4.1, 5.2, 5.5, 4.7, 5.7, 3.5, 5.0, 5.4, 5.3, 4.9],
    'GDP()': [1.05, 6.5, 4.2, 4.6, 3.9, 4.0, 3.1, 4.6, 5.4, 1.1],
    '': ['', '', '', '', '', '', '', '', '', '']
})

# 1
plt.figure(figsize=(12, 8))
colors = {'': 'red', '': 'blue', '': 'green', '': 'purple', '': 'orange', '': 'brown'}

scatter = plt.scatter(
    cities['GDP()'], 
    cities['(/)'], 
    s=cities['()']/10,  # 
    c=cities[''].map(colors),
    alpha=0.7,
    edgecolors='w', 
    linewidth=1
)

# 
plt.title('GDP')
plt.xlabel('GDP()')
plt.ylabel('(/)')

# 
legend1 = plt.legend(
    handles=scatter.legend_elements()[0], 
    labels=['', '', '', '', '', ''],
    title='',
    loc='upper left'
)
plt.gca().add_artist(legend1)

# 
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)
legend2 = plt.legend(handles, labels, title="()", loc='lower right', 
                    labelspacing=1.5, borderpad=1.5)

# 
for i, row in cities.iterrows():
    plt.annotate(
        row[''],
        (row['GDP()'], row['(/)']),
        fontsize=10,
        xytext=(5, 5),
        textcoords='offset points'
    )

plt.grid(True, linestyle='--', alpha=0.7)

# 2
plt.figure(figsize=(12, 8))
sns.scatterplot(
    x='(L/100km)', 
    y='(kW)', 
    hue='',
    size='()',
    sizes=(100, 600),  # 
    alpha=0.7,
    palette='viridis',
    data=cars
)

plt.title('')
plt.xlabel('(L/100km)')
plt.ylabel('(kW)')
plt.grid(True, linestyle='--', alpha=0.7)

# 
for i, row in cars.iterrows():
    plt.annotate(
        row[''],
        (row['(L/100km)'], row['(kW)']),
        fontsize=9,
        xytext=(5, 5),
        textcoords='offset points'
    )

# 3
plt.figure(figsize=(12, 8))
continents = {'': 'red', '': 'blue', '': 'green', '': 'purple'}

scatter = plt.scatter(
    education['(%)'], 
    education['GDP(%)'], 
    s=education['GDP()']*100,  # 
    c=education[''].map(continents),
    alpha=0.7,
    edgecolors='w', 
    linewidth=1
)

plt.title('GDP')
plt.xlabel('(%)')
plt.ylabel('GDP(%)')

# 
legend1 = plt.legend(
    handles=scatter.legend_elements()[0], 
    labels=['', '', '', ''],
    title='',
    loc='upper left'
)
plt.gca().add_artist(legend1)

# 
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)
labels = ['1', '2', '3', '4', '5', '6']  # 
legend2 = plt.legend(handles, labels, title="GDP()", loc='lower right')

# 
for i, row in education.iterrows():
    plt.annotate(
        row[''],
        (row['(%)'], row['GDP(%)']),
        fontsize=10,
        xytext=(5, 5),
        textcoords='offset points'
    )

plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd
import matplotlib.colors as mcolors

# 
plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]
plt.rcParams["axes.unicode_minus"] = False  # 

# 1
cities = {
    '': (116.4074, 39.9042),
    '': (121.4737, 31.2304),
    '': (113.2644, 23.1291),
    '': (114.0579, 22.5431),
    '': (104.0658, 30.5728),
    '': (114.3055, 30.5928),
    '': (108.9480, 34.2631)
}

# 
events = [
    {'time': 0, 'city': '', 'intensity': 8},
    {'time': 10, 'city': '', 'intensity': 6},
    {'time': 20, 'city': '', 'intensity': 7},
    {'time': 30, 'city': '', 'intensity': 5},
    {'time': 40, 'city': '', 'intensity': 9}
]

# 2
# 
np.random.seed(42)
n_nodes = 30
nodes = pd.DataFrame({
    'x': np.random.rand(n_nodes),
    'y': np.random.rand(n_nodes),
    'influence': np.random.uniform(1, 5, n_nodes),  # 
    'category': np.random.choice(['A', 'B', 'C', 'D'], n_nodes)
})

# 
hotspots = [
    {'time': 0, 'node_id': 5, 'intensity': 10},
    {'time': 15, 'node_id': 12, 'intensity': 8},
    {'time': 30, 'node_id': 22, 'intensity': 7}
]

# 3
# 
stocks = {
    'A': (0.2, 0.3),
    'B': (0.4, 0.1),
    'C': (0.6, 0.2),
    'D': (0.8, 0.4),
    'E': (0.3, 0.6),
    'F': (0.5, 0.8),
    'G': (0.7, 0.5),
    'H': (0.1, 0.7)
}

# 
market_events = [
    {'time': 0, 'stock': 'A', 'intensity': 9, 'color': 'red'},
    {'time': 12, 'stock': 'E', 'intensity': 7, 'color': 'blue'},
    {'time': 25, 'stock': 'G', 'intensity': 8, 'color': 'green'},
    {'time': 38, 'stock': 'C', 'intensity': 6, 'color': 'purple'}
]

# 1
fig1, ax1 = plt.subplots(figsize=(10, 8))

# 
ax1.set_xlim(100, 130)
ax1.set_ylim(20, 45)

# 
city_points = []
for city, (lon, lat) in cities.items():
    point, = ax1.plot(lon, lat, 'o', markersize=8, color='black', label=city)
    city_points.append(point)
    ax1.annotate(city, (lon, lat), xytext=(5, 5), textcoords='offset points')

# 
ax1.set_title('')
ax1.set_xlabel('')
ax1.set_ylabel('')

# 
ax1.grid(True, linestyle='--', alpha=0.5)

# 
ripples = []

def init1():
    return []

def update1(frame):
    # 
    for ripple in ripples[:]:
        if ripple[0].get_alpha() <= 0:
            for artist in ripple:
                artist.remove()
            ripples.remove(ripple)
    
    # 
    for event in events:
        if frame >= event['time'] and frame < event['time'] + 40:
            elapsed = frame - event['time']
            if elapsed % 5 == 0:  # 5
                lon, lat = cities[event['city']]
                radius = elapsed / 4
                alpha = 0.8 - (elapsed / 40) * 0.8
                
                # 
                circle1 = plt.Circle((lon, lat), radius, fill=False, color='red', alpha=alpha, linewidth=2)
                circle2 = plt.Circle((lon, lat), radius*0.7, fill=False, color='red', alpha=alpha*0.7, linewidth=1.5)
                circle3 = plt.Circle((lon, lat), radius*0.4, fill=False, color='red', alpha=alpha*0.4, linewidth=1)
                
                ax1.add_artist(circle1)
                ax1.add_artist(circle2)
                ax1.add_artist(circle3)
                
                ripples.append([circle1, circle2, circle3])
    
    return [artist for ripple in ripples for artist in ripple]

# 
ani1 = FuncAnimation(fig1, update1, frames=np.arange(0, 60), init_func=init1, blit=True, interval=100)

# 2
fig2, ax2 = plt.subplots(figsize=(10, 8))

# 
colors = {'A': 'red', 'B': 'blue', 'C': 'green', 'D': 'purple'}
for category, color in colors.items():
    cat_nodes = nodes[nodes['category'] == category]
    ax2.scatter(cat_nodes['x'], cat_nodes['y'], c=color, s=cat_nodes['influence']*15, 
               alpha=0.7, label=category, edgecolors='w', linewidth=1)

# 
ax2.set_title('')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.legend(title='')

# 
ax2.set_xticks([])
ax2.set_yticks([])

# 
ax2.grid(True, linestyle='--', alpha=0.3)

# 
social_ripples = []

def init2():
    return []

def update2(frame):
    # 
    for ripple in social_ripples[:]:
        if ripple[0].get_alpha() <= 0:
            for artist in ripple:
                artist.remove()
            social_ripples.remove(ripple)
    
    # 
    for hotspot in hotspots:
        if frame >= hotspot['time'] and frame < hotspot['time'] + 30:
            elapsed = frame - hotspot['time']
            if elapsed % 4 == 0:  # 4
                node = nodes.iloc[hotspot['node_id']]
                x, y = node['x'], node['y']
                intensity = hotspot['intensity']
                radius = elapsed / 5
                alpha = 0.9 - (elapsed / 30) * 0.9
                
                # 
                circle1 = plt.Circle((x, y), radius, fill=False, color='orange', alpha=alpha, linewidth=2)
                circle2 = plt.Circle((x, y), radius*0.7, fill=False, color='orange', alpha=alpha*0.7, linewidth=1.5)
                
                ax2.add_artist(circle1)
                ax2.add_artist(circle2)
                
                social_ripples.append([circle1, circle2])
    
    return [artist for ripple in social_ripples for artist in ripple]

# 
ani2 = FuncAnimation(fig2, update2, frames=np.arange(0, 50), init_func=init2, blit=True, interval=150)

# 3
fig3, ax3 = plt.subplots(figsize=(10, 8))

# 
stock_points = []
for stock, (x, y) in stocks.items():
    point, = ax3.plot(x, y, 'o', markersize=8, color='black', label=stock)
    stock_points.append(point)
    ax3.annotate(stock, (x, y), xytext=(5, 5), textcoords='offset points')

# 
ax3.set_title('')
ax3.set_xlabel('')
ax3.set_ylabel('')

# 
ax3.grid(True, linestyle='--', alpha=0.5)

# 
ax3.set_xlim(0, 1)
ax3.set_ylim(0, 1)

# 
market_ripples = []

def init3():
    return []

def update3(frame):
    # 
    for ripple in market_ripples[:]:
        if ripple[0].get_alpha() <= 0:
            for artist in ripple:
                artist.remove()
            market_ripples.remove(ripple)
    
    # 
    for event in market_events:
        if frame >= event['time'] and frame < event['time'] + 35:
            elapsed = frame - event['time']
            if elapsed % 5 == 0:  # 5
                x, y = stocks[event['stock']]
                intensity = event['intensity']
                radius = (elapsed / 35) * 0.3
                alpha = 0.8 - (elapsed / 35) * 0.8
                
                # 
                circle1 = plt.Circle((x, y), radius, fill=False, color=event['color'], alpha=alpha, linewidth=2)
                circle2 = plt.Circle((x, y), radius*0.6, fill=False, color=event['color'], alpha=alpha*0.7, linewidth=1.5)
                
                ax3.add_artist(circle1)
                ax3.add_artist(circle2)
                
                market_ripples.append([circle1, circle2])
    
    return [artist for ripple in market_ripples for artist in ripple]

# 
ani3 = FuncAnimation(fig3, update3, frames=np.arange(0, 50), init_func=init3, blit=True, interval=120)

plt.tight_layout()
plt.show()
```
